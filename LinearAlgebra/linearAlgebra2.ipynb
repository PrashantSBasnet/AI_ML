```python
import numpy as np
```

```python
v =np.array([3,1]) #vector v
v
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_vectors(vectors, colors):
  plt.figure()
  plt.axvline(x=0, color='lightgray')
  plt.axhline(y=0, color='lightgray')

  for i in range(len(vectors)):
    x=np.concatenate([[0,0],vectors[i]])
    plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
      angles='xy', scale_units='xy', scale=1, color=colors[i],)
```

```python
plot_vectors([v], ['lightblue'])
plt.xlim(-1, 5)
_ = plt.ylim(-1, 5)
```

Affine Transformation

```python
A = np.array([[-1,4],[2,-2]])
A
```

```python
Av = np.dot(A,v) #applying matrix A on v
Av
```

```python
plot_vectors([v,Av],['lightblue', 'blue'])
plt.xlim(-1,5)
_ = plt.ylim(-1,5)
```

```python
#another example of applying matrix A on v2
v2= np.array([2,1])
plot_vectors([v2,Av],['lightgreen', 'green'])
plt.xlim(-1,5)
_ = plt.ylim(-1,5)
```
